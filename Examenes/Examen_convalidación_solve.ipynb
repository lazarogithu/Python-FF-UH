1.2- Implemente el algoritmo de ordenamiento **Bubble Sort**. Este algoritmo compara cada elemento de la lista con el siguiente, intercambiándolos si están en el orden incorrecto. 

```python
#Respuesta 1.1
def bubble_sort(lista):
    for i in range(len(lista)):
        for j in range(len(lista) - 1):
            if lista[j] > lista[j + 1]:
                lista[j], lista[j + 1] = lista[j + 1], lista[j]
    return lista

numeros = [64, 34, 25, 12, 22, 11, 90]
print(bubble_sort(numeros))
```

1.3- Implemente un programa que tome una cadena y devuelva el recuento de cada carácter en la cadena.

```python
cadena = "hola mundo"
caracteres = "abcdefghijklmnopqrstuvwxyz "
recuento = [0]*27  # 26 letras más el espacio

for caracter in cadena:
    for i, letra in enumerate(caracteres):
        if caracter == letra:
            recuento[i] += 1

for i, letra in enumerate(caracteres):
    print(f"'{letra}': {recuento[i]}")
```

#### 2- Estructuras de datos. Funciones

2.1- Implemente una función que tome una lista de números y devuelva dos nuevas listas que contengan solo los números pares e impares respectivamente de la lista original.

```python
def separar_pares_impares(numeros):
    pares = [num for num in numeros if num % 2 == 0]
    impares = [num for num in numeros if num % 2 != 0]
    return pares, impares

pares, impares = separar_pares_impares([1, 2, 3, 4, 5, 6])
print("Pares: ", pares)
print("Impares: ", impares) 
```

2.2- Dada una lista de tuplas, ordene la lista en base al segundo elemento de cada tupla.
 Por ejemplo, dada la lista [(1, 2), (3, 3), (1, 1)], el resultado debería ser [(1, 1), (1, 2), (3, 3)].


```python
def ordenar_por_segundo_elemento(tuplas):
    return sorted(tuplas, key=lambda x: x[1])

print(ordenar_por_segundo_elemento([(1, 2), (3, 3), (1, 1)]))
```

2.3-  Dado un diccionario, invierta las llaves y los valores utilizando comprehensions. Por ejemplo, dado el diccionario {'a': 1, 'b': 2}, el resultado debería ser {1: 'a', 2: 'b'}.


```python
def invertir_diccionario(diccionario):
    return {v: k for k, v in diccionario.items()}

print(invertir_diccionario({'a': 1, 'b': 2, 3:'l'}))
```

4-El movimiento de dos masas unidas por un resorte se puede modelar con el siguiente sistema de ecuaciones diferenciales ordinarias:



            dx1/dt = y1
            dy1/dt = -k*(x1 - x2)
            dx2/dt = y2
            dy2/dt = -k*(x2 - x1)
donde x1 y x2 son las posiciones de las masas, y1 y y2 son sus velocidades, y k es la constante del resorte.

-Resuelva este sistema de EDO para k=0.3, condiciones iniciales x1=1, y1=0, x2=-1, y2=0, en un intervalo de tiempo de 0 a 50.



-Visualize las posiciones de las masas en función del tiempo en dos gráficos separados.



En cada gráfico, incluye una línea horizontal punteada en y=0 para representar el punto de equilibrio.
Asegúrate de que los gráficos tengan títulos, etiquetas de los ejes y leyendas.

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

#función que representa el sistema de EDO
def modelo(z, t):
    k = 0.3
    dx1dt = z[1]
    dy1dt = -k * (z[0] - z[2])
    dx2dt = z[3]
    dy2dt = -k * (z[2] - z[0])
    dzdt = [dx1dt, dy1dt, dx2dt, dy2dt]
    return dzdt

#condiciones iniciales
z0 = [1, 0, -1, 0]

# tiempo de integración de 0 a 50 unidades de tiempo
t = np.linspace(0,50)

# sol del sistema de EDO
z = odeint(modelo, z0, t)

#visualizacion
plt.figure(figsize=(12,6))

plt.subplot(211)
plt.plot(t, z[:, 0], label='x1(t)')
plt.axhline(0, color='gray', linestyle='--')
plt.xlabel('tiempo')
plt.ylabel('posición')
plt.title('Posición de la masa 1 en función del tiempo')
plt.legend()

plt.subplot(212)
plt.plot(t, z[:, 2], label='x2(t)', color='orange')
plt.axhline(0, color='gray', linestyle='--')
plt.xlabel('tiempo')
plt.ylabel('posición')
plt.title('Posición de la masa 2 en función del tiempo')
plt.legend()

plt.tight_layout()
plt.show()
```
